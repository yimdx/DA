```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("human_feedback.csv")
print(data)
```

```python
chosen = data["chosen_layout"]
count = data.groupby("chosen_layout").count()
print(count)
```

```python
plt.figure(figsize=(10, 6))
count["graph_id"].plot(kind='bar', color='skyblue')
plt.title("Count of Each Chosen Layout")
plt.xlabel("Chosen Layout")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
