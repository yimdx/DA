From efficient and high quality force-directed graph drawing.

### Octree
this is to reduce complexity.

- An octree data structure is constructed by first forming a square (or cube in 3D) that encloses all vertices.This is the level 0 square. This square is subdivided into 4 squares (or 8 cubes) if it contains more than l vertex, and forms the level 1 squares. This processis repeated until level L, where each square contains no more than 1 vertex.

- supernodes
  - the center of gravity of a square.
- how far
  - distance compared to the square length.
- complexity(O(VlogV))
- max tree level(limit the number of levels of octree)


### Mutilevel
this is to get a global minimum.
#### Three phase
- Coarsening
  - the coarser graph encapsulate the information needed to layout its parent
- Coarsest Graph Layout(On the top level, use force directed algorithm to get the best layout)
- Prolongation and Refinement
   - The graph layouts are recursively refined at each level, starting from coarser graphs and moving towards finer graphs.
   - If a graph $ G^{i+1} $ is derived from $ G^i $ using edge contraction (EC), the position of a vertex $ u \in V^{i+1} $ is inherited from its collapsed vertices $ v, w \in V^i $.
   - When using maximal independent vertex sets (MIVS), a vertex in $ V^i $ inherits its position directly if it belongs to the MIVS, or its position is averaged based on its neighbors in the MIVS.
   - a conservative step-length update scheme.
   - scaling
   - only contributions from nearby vertices within a radius are considered.
